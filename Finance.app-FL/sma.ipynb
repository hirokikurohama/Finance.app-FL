```python
import pandas as pd
import numpy as np
import talib as ta
from pandas_datareader import data
import matplotlib.pyplot as plt
%matplotlib inline
import warnings
warnings.simplefilter('ignore')
```

```python
start = '2021-08-31'
end = '2022-08-31'
company_code = '^N225'
```

```python
df = data.DataReader(company_code, 'yahoo', start, end)
date=df.index
close=df['Adj Close']

span1=5
span2=25
span3=50
span12=12
span26=26


df['sma1'] = close.rolling(window=span1).mean()
df['sma2'] = close.rolling(window=span2).mean()
df['sma3'] = close.rolling(window=span3).mean()
df['macd'], df['macdsignal'], df['macdhist'], = ta.MACD(close, fastperiod=12, slowperiod=26, signalperiod=9)
#df['sma12'] = close.rolling(window=span12).mean()
#df['sma26'] = close.rolling(window=span26).mean()
#df['ema12'] = close.rolling(window=span12).mean()
#df['ema26'] = close.rolling(window=span26).mean()



plt.figure(figsize=(30,15))
plt.subplot(2,1,1) #,,
plt.plot(date,close,label='Close')
plt.plot(date,df['sma1'],label='sma1')
plt.plot(date,df['sma2'],label='sma2')
plt.plot(date,df['sma3'],label='sma3')

plt.legend() #

plt.subplot(2,1,2)
plt.fill_between(date, df['macdhist'], color ='gray', alpha=0.5, label='MACD')
plt.hlines(0,start,end,"gray",linestyles="dashed")
plt.plot(date,df['macd'],label='macd')
plt.plot(date,df['macdsignl'],label='macdsignal')
plt.legend()
```
